```python
import matplotlib.pyplot as plt
import numpy as np
from math import log
```

```python
def gradient_descent(derivative_func, learning_rate=0.02, 
                     initial_value=0.5, precision=0.001, max_iter=300, print_steps=False, steps = 0):
    t=initial_value
    t_list=[]

    for i in range(1,max_iter+1):
        t_list.append(t)
        t = t - learning_rate * derivative_func(t)
        
        if abs(learning_rate * derivative_func(t))<=precision:
        #if derivative_func(x)==0:
            break
        if print_steps and i<steps+1 :
            print(t,derivative_func(t), learning_rate*derivative_func(t),t_list)
    return t,t_list,i
```
